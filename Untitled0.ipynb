```python
import pandas as pd
import json
import colour
```

```python
def load_data(file_name):
    header_file = '../data/{}.txt'.format(file_name)
    header = !cut -f1 $header_file
    data = pd.read_csv('../data/{}.csv'.format(file_name), sep=';',encoding='utf-8', header=None)
    data.columns = header + ['noname_{}'.format(i) for i in range(len(data.columns) - len(header))]
    
    return data

ct = pd.read_csv('ct.csv', encoding='utf-8')

obras = load_data('../copa2014/data/20140620_Empreendimento')
cidades = load_data('../copa2014/data/20140620_CidadeSede')
obras_tema = load_data('../copa2014/data/20140620_Tema')

```

```python
cid_obra = []

total = ct['Total Previsto'].sum()

for cid, cid_obras in obras[(obras['IdTema'] == 6) | (obras['IdTema'] == 8)].groupby('IdCidadeSede'):
    if len(cidades[cidades['IdCidadeSede'] == int(cid)]['DescCidadeSede']) > 0:
        cidade = cidades[cidades['IdCidadeSede'] == int(cid)]['DescCidadeSede'].values[0]
        cid_obra.append({'label': cidade.encode('utf-8'), 
                         'amount': cid_obras['ValTotalPrevisto'].sum(), 
                         'color': colour.Color(hsl=(float(cid_obras['ValTotalPrevisto'].sum())/(total), 0.7, 0.5)).hex,
                         'children':[]})
        for idx, obra in cid_obras.iterrows():
            cid_obra[-1]['children'].append({'label': obra['DescEmpreendimento'].encode('utf-8'), 'amount': obra['ValTotalPrevisto']})
                             
                             
                             
                             
```

```python
d = {'label': 'Total gasto em Estádios e Mobilidade'
    ,'amount': total
    ,'children': []}

d['children'] = cid_obra
```

```python
f = json.dumps(d, ensure_ascii=False)
ff = open('j', 'w')
ff.write(f)
ff.flush()
```
